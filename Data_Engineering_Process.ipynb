# **Data Engineering Process**

There are several steps in Data Engineering process.

1.  **Extract** - Data extraction is getting data from multiple sources. Ex. Data extraction from a website using Web scraping or gathering information from the data that are stored in different formats(JSON, CSV, XLSX etc.).

2.  **Transform** - Transforming the data means removing the data that we don't need for further analysis and converting the data in the format that all the data from the multiple sources is in the same format.

3.  **Load** - Loading the data inside a data warehouse. Data warehouse essentially contains large volumes of data that are accessed to gather insights.

## **2. JSON file Format**

JSON (JavaScript Object Notation) adalah format pertukaran data yang ringan. Mudah untuk dibaca dan ditulis oleh manusia.

JSON adalah format data yang independen dari bahasa pemrograman. JSON berasal dari JavaScript, tetapi banyak bahasa pemrograman modern yang menyertakan kode untuk menghasilkan dan mengurai data dalam format JSON. Ini adalah format data yang sangat umum dengan berbagai aplikasi.

Teks dalam JSON dituliskan melalui string yang diapit oleh tanda kutip dan berisi nilai-nilai dalam pemetaan kunci-nilai dalam { }. Ini mirip dengan kamus (dictionary) dalam Python.

```python
import json
```

**Import JSON to a File**

```python
import json
person = {
    'first_name' : 'Mark',
    'last_name' : 'abc',
    'age' : 27,
    'address': {
        "streetAddress": "21 2nd Street",
        "city": "New York",
        "state": "NY",
        "postalCode": "10021-3100"
    }
}
```

**Serialization using dump() funtion**

```python
# Serializing json
json_object = json.dumps(person, indent = 4)

# Writing to sample.json
with open("sample.json", "w") as outfile:
    outfile.write(json_object)
```

```python
print(json_object)
```

### **Reading JSON to a File**

**Using Json.load()**

```python
import json

# Opening JSON file
with open('sample.json', 'r') as openfile:

    # Reading from json file
    json_object = json.load(openfile)

print(json_object)
print(type(json_object))
```

**Table Pandas**

```python
import json
import pandas as pd

# Opening JSON file
with open('sample.json', 'r') as openfile:
    # Reading from json file
    json_object = json.load(openfile)

df = pd.DataFrame(json_object)
df
```

## **3. XML file format**

XML (Extensible Markup Language) juga dikenal sebagai Bahasa Markup yang dapat Diperluas. Sesuai dengan namanya, ini adalah bahasa markup. Ini memiliki aturan tertentu untuk encoding data. Format file XML adalah format file yang dapat dibaca oleh manusia dan mesin.

Pandas tidak menyertakan metode apa pun untuk membaca dan menulis file XML. Di sini, kita akan melihat bagaimana kita dapat menggunakan modul lain untuk membaca data dari file XML, dan memuatnya ke dalam DataFrame Pandas.

**Writing with xml.etree.ElementTree**

```python
import xml.etree.ElementTree as ET

# create the file structure
employee = ET.Element('employee')
details = ET.SubElement(employee, 'details')
first = ET.SubElement(details, 'firstname')
second = ET.SubElement(details, 'lastname')
third = ET.SubElement(details, 'age')
first.text = 'Shiv'
second.text = 'Mishra'
third.text = '23'

# create a new XML file with the results
mydata1 = ET.ElementTree(employee)
# myfile = open("items2.xml", "wb")
# myfile.write(mydata)
with open("new_sample.xml", "wb") as files:
    mydata1.write(files)
```

```python
import pandas as pd

import xml.etree.ElementTree as etree

!curl -o Sample-employee-XML-file.xml https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%205/data/Sample-employee-XML-file.xml
```

```python
tree = etree.parse("Sample-employee-XML-file.xml")

root = tree.getroot()
columns = ["firstname", "lastname", "title", "division", "building","room"]

df = pd.DataFrame(columns = columns)

for node in root:

    firstname = node.find("firstname").text

    lastname = node.find("lastname").text

    title = node.find("title").text

    division = node.find("division").text

    building = node.find("building").text

    room = node.find("room").text

    df = df._append(pd.Series([firstname, lastname, title, division, building, room], index = columns), ignore_index = True)
```

```python
df
```

**Reading XML Using Pandas**

```python
# Herein xpath we mention the set of xml nodes to be considered for migrating  to the dataframe which in this case is details node under employees.
df=pd.read_xml("Sample-employee-XML-file.xml", xpath="/employees/details")
```

**Ubah Data Jadi CSV**

```python
df.to_csv("employee.csv", index=False)
```
